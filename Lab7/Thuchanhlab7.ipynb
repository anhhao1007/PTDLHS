4. Loại bỏ stopwords khỏi văn bản

```python
import nltk
nltk.download('punkt')
nltk.download('stopwords')

from nltk.corpus import stopwords

text = "This is a simple example to demonstrate removing stopwords from a sentence."

# Tokenize
tokens = nltk.word_tokenize(text)

# Stopwords tiếng Anh
sw = set(stopwords.words('english'))

# Lọc
filtered = [w for w in tokens if w.lower() not in sw]

print("Trước:", tokens)
print("Sau:", filtered)
```

5. Bỏ qua stopwords từ danh sách stopwords (lọc nâng cao)

```python
from nltk.corpus import stopwords

text = "This is a simple example showing how to ignore stopwords selectively."

tokens = nltk.word_tokenize(text)

sw = set(stopwords.words('english'))

# Giữ lại một số từ dù là stopword
keep_words = {'not', 'no'}

filtered = [
    w for w in tokens
    if (w.lower() not in sw) or (w.lower() in keep_words)
]

print(filtered)
```

6. Tìm định nghĩa + ví dụ của từ bằng WordNet

```python
import nltk
nltk.download('wordnet')
nltk.download('omw-1.4')

from nltk.corpus import wordnet as wn

word = "computer"

synsets = wn.synsets(word)

for syn in synsets[:3]:  # lấy 3 nghĩa đầu
    print("Định nghĩa:", syn.definition())
    print("Ví dụ:", syn.examples())
    print()
```

7. Tìm từ đồng nghĩa & trái nghĩa

```python
from nltk.corpus import wordnet as wn

word = "good"

synonyms = set()
antonyms = set()

for syn in wn.synsets(word):
    for lemma in syn.lemmas():
        synonyms.add(lemma.name())
        if lemma.antonyms():
            antonyms.add(lemma.antonyms()[0].name())

print("Đồng nghĩa:", synonyms)
print("Trái nghĩa:", antonyms)
```

8. Xem tag + regex

```python
import nltk

# Tải dữ liệu
nltk.download('brown')
nltk.download('universal_tagset')

from nltk.corpus import brown

# Lấy tag
tags = brown.tagged_words(tagset='universal')

print("Một số tag:", tags[:20])

# Lọc danh từ
nouns = [word for (word, tag) in tags if tag == 'NOUN']
print("Danh từ:", nouns[:10])
```

9. So sánh độ giống nhau của 2 danh từ

```python
from nltk.corpus import wordnet as wn

w1 = wn.synset('dog.n.01')
w2 = wn.synset('cat.n.01')

print("Độ giống nhau:", w1.wup_similarity(w2))
```

10. So sánh 2 động từ

```python
from nltk.corpus import wordnet as wn

v1 = wn.synset('run.v.01')
v2 = wn.synset('walk.v.01')

print("Độ giống nhau:", v1.wup_similarity(v2))
```

11. Đếm tên nam/nữ + in 10 tên đầu

```python
import nltk
nltk.download('names')

from nltk.corpus import names

male = names.words('male.txt')
female = names.words('female.txt')

print("Số tên nam:", len(male))
print("Số tên nữ:", len(female))

print("10 tên nam:", male[:10])
print("10 tên nữ:", female[:10])
```

12. 15 tên ngẫu nhiên có nhãn

```python
import random
from nltk.corpus import names

labeled_names = ([(name, 'male') for name in names.words('male.txt')] +
                 [(name, 'female') for name in names.words('female.txt')])

random.shuffle(labeled_names)

print(labeled_names[:15])
```

13. Lấy ký tự cuối + nhãn

```python
from nltk.corpus import names

labeled_names = ([(name, 'male') for name in names.words('male.txt')] +
                 [(name, 'female') for name in names.words('female.txt')])

features = [(name[-1], gender) for (name, gender) in labeled_names]

print(features[:10])
```
