```python
import nltk

# Tải dữ liệu cần thiết
nltk.download('gutenberg')
nltk.download('punkt')
nltk.download('stopwords')

from nltk.corpus import gutenberg, stopwords
import string

# Lấy nội dung Macbeth
macbeth = gutenberg.words('shakespeare-macbeth.txt')

# Stopwords
sw = stopwords.words('english')

# Dấu câu
punctuation = set(string.punctuation)

# Lọc từ
macbeth_filtered2 = [
    w.lower()
    for w in macbeth
    if w.lower() not in sw and w.lower() not in punctuation
]

# Tần suất
fd = nltk.FreqDist(macbeth_filtered2)
print(fd.most_common(10))
```

```python
import nltk
import string
from nltk.corpus import stopwords
from urllib import request

# Tải dữ liệu tokenizer
nltk.download('punkt')
nltk.download('punkt_tab')  # FIX lỗi
nltk.download('stopwords')

# Tải text
url = "http://www.gutenberg.org/files/2554/2554-0.txt"
raw = request.urlopen(url).read().decode('utf8')

# Làm sạch
start = raw.find("*** START OF THE PROJECT GUTENBERG EBOOK")
end = raw.find("*** END OF THE PROJECT GUTENBERG EBOOK")
clean_raw = raw[start:end]

# Tokenize
tokens = nltk.word_tokenize(clean_raw)

# Stopwords + punctuation
sw = stopwords.words('english')
punctuation = set(string.punctuation)

# Lọc từ
filtered = [
    w.lower() for w in tokens
    if w.lower() not in sw and w not in punctuation
]

# Đếm tần suất
fd = nltk.FreqDist(filtered)
print(fd.most_common(10))
```

7. Rút trích văn bản từ trang html

```python
# 1. Import thư viện
import nltk
import string
from urllib import request
from bs4 import BeautifulSoup
from nltk.corpus import stopwords

# 2. Tải dữ liệu cần thiết
nltk.download('punkt')
nltk.download('punkt_tab')   # tránh lỗi tokenizer
nltk.download('stopwords')

# 3. Lấy HTML từ web
url = "http://news.bbc.co.uk/2/hi/health/2284783.stm"
html = request.urlopen(url).read().decode('utf8')

# 4. Parse HTML bằng BeautifulSoup
soup = BeautifulSoup(html, 'html.parser')

# 5. Lấy nội dung chính (chỉ lấy thẻ <p>)
paragraphs = soup.find_all('p')
text = ' '.join(p.get_text() for p in paragraphs)

# 6. Tokenize
tokens = nltk.word_tokenize(text)

# 7. Stopwords + dấu câu
sw = stopwords.words('english')
punctuation = set(string.punctuation)

# 8. Lọc từ
filtered = [
    w.lower() for w in tokens
    if w.lower() not in sw and w not in punctuation
]

# 9. Đếm tần suất
fd = nltk.FreqDist(filtered)

# 10. In kết quả
print("Top 10 từ xuất hiện nhiều nhất:")
print(fd.most_common(10))
```

8. Phân tích cảm xúc người dùng

```python
nltk.download('movie_reviews')
```

```python
import random
reviews = nltk.corpus.movie_reviews
documents = [(list(reviews.words(fileid)), category)
for category in reviews.categories()
for fileid in reviews.fileids(category)]
random.shuffle(documents)
```

```python
first_review = ' '.join(documents[0][0])
print(first_review)
```

```python
documents[0][1]
```

```python
all_words = nltk.FreqDist(w.lower() for w in reviews.words())
word_features = list(all_words)
```

```python
def document_features(document, word_features):
    document_words = set(document)
    features = {}

    for word in word_features:
        features[word] = (word in document_words)

    return features
```

```python
featuresets = [
    (document_features(d, word_features), c)
    for (d, c) in documents
]

len(featuresets)
```

```python
train_set, test_set = featuresets[1500:], featuresets[:500]
classifier = nltk.NaiveBayesClassifier.train(train_set)
```

```python
import random

random.shuffle(featuresets)

train_set, test_set = featuresets[1500:], featuresets[:500]

classifier = nltk.NaiveBayesClassifier.train(train_set)

print("Accuracy:", nltk.classify.accuracy(classifier, test_set))
```
